```python
import numpy as np 
import pandas as pd
import matplotlib.pylab as plt
%matplotlib inline

Data = pd.read_csv('C://Users//knguy//Desktop//School//Spring 2018//SaberMetrics//lahman1990-2017Project.csv')
```

# Here you can input the different ways to filter the data, you can choose by playerID, FirstName, LastName, Team, or Year. The Stat input is so that you can see the top performers in that statistic.The AtBatsMin is an input to choose the minimum of at bats you want to filter by. 

```python
playerID = 'stantmi03'
FirstName = 'Nolan'
LastName = 'Arenado'
teamID = 'COL'
#Data only uses from year 1990 - 2017
yearID = 2017
#The different stats you can choose from are HPerSO, HRPerSO, RBIPerSO, RPerSO KPerAB, and BA
Stat = 'HPerSO'
AtBatsMin = 200
```

# This will filter by PlayerID

```python
SearchPlayerID = Data[Data['playerID'] == playerID]
SearchPlayerID
```

# This will filter by First and Last Name

```python
SearchPlayerFirstName = Data[Data['FirstName'] == FirstName]
SearchPlayerFull = SearchPlayerFirstName[SearchPlayerFirstName['LastName'] == LastName]
SearchPlayerFull
```

# This will filter by TeamID and YearID

```python
SearchTeam = Data[Data['teamID'] == teamID]
SearchTeamYear = SearchTeam[SearchTeam['yearID'] == yearID]
SearchTeamYear = SearchTeamYear[SearchTeamYear.AB != 0]
SearchTeamYear.sort_values(Stat, ascending  = False)
```

# This will filter by YearID and the Stat you have chosen

```python
SearchYear = Data[Data['yearID'] == yearID]
SearchYear = SearchYear[SearchYear.AB > AtBatsMin]
SearchYear.sort_values(Stat, ascending  = False)
```

# Below is the averages for the chosen year for each stat

```python
SearchYear.mean()
```

# This is a correlation between BA and HperSO 

```python
Data['HPerSO'].corr(Data['BA'], method='pearson')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(6,6))
axes.scatter(Data["HPerSO"], Data["BA"], color = "steelblue")
axes.grid(alpha=0.25)
axes.set_xlabel("Hits Per SO", fontsize=16)
axes.set_ylabel("Batting Average", fontsize=16)
axes.set_title("Hits Per SO vs Batting Average", fontsize=16)
```
